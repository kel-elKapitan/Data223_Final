```python
import boto3
import pandas as pd

s3 = boto3.client('s3')
bucket_name = 'data-eng-223-final-project'
data_ls = []

try:
    response = s3.list_objects_v2(Bucket=bucket_name)
    
    # Check if any objects exist in the bucket
    if 'Contents' in response:
        objects = response['Contents']
        
        # Iterate over the objects and print their keys
        for obj in objects:
            object_key = obj['Key']
            if object_key.startswith('Academy/Data'):
                # print(object_key)
                data_ls.append(object_key)
    else:
        print('No objects found in the bucket.')
        
except Exception as e:
    print(f"Error listing objects: {e}")

rows = 0
for obj_name in data_ls:
    file_name = obj_name.split('/')[-1]  # Extract the file name from the object key
    print(file_name)
    try:
        response = s3.get_object(Bucket=bucket_name, Key=obj_name)
        df = pd.read_csv(response['Body'])
        print(df.columns)
        display(df)
        
        rows+=df.shape[0]
        # Use the DataFrame 'df' for further processing or analysis
    except Exception as e:
        print(f"Error reading file {file_name}: {e}")
print(rows)
```
